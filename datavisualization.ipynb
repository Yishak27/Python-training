Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
        'Transaction Amount': [20000, 25000, 22000, 27000, 28000, 30000, 31000, 35000, 33000, 36000, 34000, 37000]}

print(pd.DataFrame(data).head(3))
df = pd.DataFrame(data)
plt.plot(df['Month'], df['Transaction Amount'], marker='o', linestyle='-', color='b')
plt.xlabel('TA in$')
plt.ylabel("Month")
plt.title('Transaction graph')
plt.grid()

```

```python
plt.bar(df["Month"], df["Transaction Amount"],align='center')
```

```python
plt.pie(df["Transaction Amount"],data=df,labels=df["Month"],normalize='true',radius=1,textprops=None)
plt.show()
```

```python
import numpy as np
# Simulate some transaction amount data
transaction_amounts = np.random.normal(1000, 200, 1000)  # 1000 transactions with a mean of $1000

# Plotting
plt.hist(transaction_amounts, bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Daily Transaction Amounts')
plt.xlabel('Transaction Amount ($)')
plt.ylabel('Frequency')
plt.show()
```
