Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
        'Transaction Amount': [20000, 25000, 22000, 27000, 28000, 30000, 31000, 35000, 33000, 36000, 34000, 37000]}

print(pd.DataFrame(data).head(3))
df = pd.DataFrame(data)
plt.plot(df['Month'], df['Transaction Amount'], marker='o', linestyle='-', color='b')
plt.xlabel('TA in$')
plt.ylabel("Month")
plt.title('Transaction graph')
plt.grid()

```

```python
plt.bar(df["Month"], df["Transaction Amount"],align='center')
```

```python
plt.pie(df["Transaction Amount"],data=df,labels=df["Month"],normalize='true',radius=1,textprops=None)
plt.show()
```

```python
import numpy as np
# Simulate some transaction amount data
transaction_amounts = np.random.normal(1000, 500, 1000)  # 1000 transactions with a mean of $1000

# Plotting
plt.hist(transaction_amounts, bins=30, color='purple', edgecolor='black')
plt.title('Distribution of Daily Transaction Amounts')
plt.xlabel('Transaction Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
# Sample data
data = {'Account Type': ['Savings', 'Savings', 'Checking', 'Checking', 'Credit', 'Credit', 'Loan', 'Loan'],
        'Transaction Amount': [2000, 303300, 1500, 1600, 7000, 7200, 50000, 50500]}
df = pd.DataFrame(data)
# Plotting
sns.boxplot(x='Account Type', y='Transaction Amount', data=df)
plt.title('Transaction Amount by Account Type')
sns.axes_style(style='dark')
plt.show()
```

```python
# Sample correlation data
data = {'ATM Transactions': [50, 39, 20], 'Online Payments': [0, 20, 75], 'In-branch Transactions': [20, 15, 50]}
df = pd.DataFrame(data, index=['Savings', 'Checking', 'Credit'])
# Plotting
sns.heatmap(df, annot=True, cmap='YlGnBu')
plt.title('Transaction Type Distribution by Account Type')
plt.xlabel('Transaction Type')
plt.ylabel('Account Type')
plt.show()
```
